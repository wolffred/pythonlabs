```python
import pandas as pd
import numpy as np
import csv
```

```python
data_16 = pd.read_csv("AAPL-2016.csv")
data_17 = pd.read_csv("AAPL-2017.csv")
```

```python
data_16.shape
data_17.shape
```

```python
data_17.drop([250], inplace=True)
data_17.tail()
```

```python
array_16 = data_16[['Open', 'Volume']].to_numpy()
array_17 = data_17[['Open', 'Volume']].to_numpy()
```

```python
sumOfArray16 = array_16.sum(axis=0) 
print("sum of volume for 2016 = ", sumOfArray16[1])
sumOfArray17 = array_17.sum(axis=0)
print("sum of volume for 2017 = ",sumOfArray17[1] )
```

Q2

```python
change = np.subtract(array_16, array_17)
print(sumOfArray16[1] - sumOfArray17[1])
change.sum(axis=0)[1] 
```

Q3

```python
print(change)
np.set_printoptions(formatter={'float': '{:6.2f}'.format})
print("########################################")
print(change)
```

Q4

```python
combined_years = np.concatenate((array_16, array_17))
```

Q5

```python
with open("AAPL_2016_2017.csv", 'w', newline="") as newfile:
    writer = csv.writer(newfile)
    writer.writerow(["Open Price", "Volume"])
    writer.writerows(combined_years)
    
```

```python
np.savetxt("AAPL_2016_2017.txt",combined_years, fmt = "%10.2f", delimiter = ",")
```
