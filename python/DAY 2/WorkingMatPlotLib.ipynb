```python
df_2016 = pd.read_csv("AAPL-2016.csv", parse_dates =True)
df_2017 = pd.read_csv("AAPL-2017.csv", parse_dates =True)
df_2017.head()
```

## Opening price for Apple in 2016 and 2017

```python
opening_16 = df_2016['Open']
opening_17 = df_2017['Open']
opening_17.drop([250], inplace = True)
opening_17.shape
plt.plot(opening_16,'.', label = "2016")
plt.plot(opening_17, '.', label = "2017")
plt.legend()
plt.savefig("plotOf2016And17.png")
plt.show()
```
