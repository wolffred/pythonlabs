```python
import numpy as np
import csv
```

READING FROM CSV FILE

```python
aapl_2016 = np.genfromtxt('AAPL-2016.csv', delimiter=',', skip_header = 1)
aapl_2017 = np.genfromtxt('AAPL-2017.csv', delimiter=',', skip_header = 1)
aapl_2016
aapl_2017
```

GETTING NUMBER OF ROWS

```python
print(aapl_2016.shape)
print(aapl_2017.shape)
```

TRIMMING AAPL_2017'S ROWS SO THEY MATCH

```python
aapl_2017 = np.delete(aapl_2017,250,0)
```

```python
print(aapl_2016.shape)
print(aapl_2017.shape)
```

 new stock prices and volume traded into another NumPy array

```python
volume_open_2016 = aapl_2016[:,[1, 6]]
volume_open_2017 = aapl_2017[:,[1, 6]]
volume_open_2016
volume_open_2017
```

SUM OF ALL STOCK PRICES AND VOLUME TRADED IN A YEAR

```python
sum_2016 = volume_open_2016.sum(axis = 0)
sum_2017 = volume_open_2017.sum(axis = 0)
sum_2016
sum_2017
```

VOLUME TRADED FOR BOTH YEARS

```python
volume_2016 = np.array(sum_2016[1], dtype = np.int64)
volume_2017 = np.array(sum_2017[1], dtype = np.int64)
```

```python
print(volume_2016)
print(volume_2017)
```

DIFFERENCE OF VOLUME TRADED BETWEEN BOTH YEARS

```python
difference = np.subtract(volume_2017, volume_2016)
difference
```

SUBTRACT COLUMNS OF BOTH ARRAYS, SUM RESULT AND DISPLAY JUST THE VOLUME TRADED

```python
change = np.subtract(volume_open_2017, volume_open_2016)
```

```python
change
```

```python
change = change.sum(axis = 0)
```

```python
change[1]
```

CONCATENATE TABLE 2017 TO THE END OF TABLE 2016

```python
combined_years = np.concatenate((volume_open_2016, volume_open_2017))
```

```python
with open("AAPL_2016_2017.csv", 'w', newline="") as newfile:
    writer = csv.writer(newfile)
    writer.writerow(["Open Price", "Volume"])
    writer.writerows(combined_years)
```

CREATE TEXT FILE WITH COMBINED YEARS

```python
np.savetxt("AAPL_2016_2017.txt",combined_years, fmt = "%10.2f", delimiter = ",")
```
