```python
volume_open_2016 = y[:,[1, 6]]
volume_open_2017 = z[:,[1, 6]]
```

```python
sum_2016 = volume_open_2016.sum(axis = 0)
sum_2017 = volume_open_2017.sum(axis = 0)
```

```python
print(sum_2017)
```

```python
new_sum_2016 = np.array(sum_2016, dtype = np.int64)
```

```python
print(new_sum_2016)
```

```python
new_sum_2017 = np.array(sum_2017, dtype = np.int64)
```

```python
print(new_sum_2017)
```

```python
combined_years = np.concatenate((volume_open_2016, volume_open_2017))
combined_years
```

```python
with open("AAPL_2016_2017.csv", 'w', newline="") as newfile:
    writer = csv.writer(newfile)
    writer.writerow(["Open Price", "Volume"])
    writer.writerows(combined_years)
    
```
